```python
import pandas as pd
import numpy as ny
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv");
crew=pd.read_csv("tmdb_5000_credits.csv");
print(movies.iloc[0])
print(crew.iloc[0])
```

```python
# what columns i want in my code 
# genres , id , overview , realease date ,title ,key words , cast crew 
# id , title , tags => [ genres , overview ,key words  ]
# merge one list into another }
movies=movies.merge(crew ,on='title');
print(movies.iloc[0])
movies=movies[['id','genres','overview','title','keywords',"crew",'cast']]
print(movies.iloc[0]['genres'])
```

```python
import ast 

movies.head()
```

```python
# now  we have to make function to extract the string value 
def convert(text):
    names=[]
    for i in ast.literal_eval(text):
        names.append(i['name'])

    return names
```

```python
# this  function is used to apply conver t function in genres 
movies['genres']=movies['genres'].apply(convert);
```

```python
# now applying same function on keywords 
movies['keywords']=movies['keywords'].apply(convert);
movies.head()
```

```python
# split the overview  and convert into list
movies['overview'] = movies['overview'].fillna('').apply(lambda x: x.split())
```

```python
# function to get starting three 3 charctewr name
def convert3(text):
    l=[];
    count=0;
    for i in ast.literal_eval(text):
        if (count<3):
            l.append(i['name'])
        count=count+1
       

    return l        
    
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.iloc[0]
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if(i['job']=='Director'):
            l.append(i['name'])
            break;
    return l
        
```

```python
# we have to extract the  name of directoppr and then we have to combine all 
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.iloc[0]
```

```python
# now we removing the extra space 
def remove_space(L):
    l1=[]
    for i in L:
        l1.append(i.replace(" ",""))
    return l1    
```

```python
movies['crew']=movies['crew'].apply(remove_space)
movies['cast']=movies['cast'].apply(remove_space)
movies['overview']=movies['overview'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head();
```

```python
movies['tags']=movies['keywords']+movies['genres']+movies['cast']+movies['crew']+movies['overview'];
```

```python
movies.iloc[0]
movies=movies.drop(columns=['keywords','genres','cast','crew','overview'])
```

```python
# convert the text in single machine model
movies['tags']=movies['tags'].apply( lambda x:" ".join(x));
movies.iloc[0]
```

```python
type(movies['tags'][0])  # list ✅
print(type(movies['tags'][0]))
print(movies['tags'][0][:100])
```

```python
 # import the sklearn count vectorizor
from sklearn.feature_extraction.text  import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
# now define the vector
vector=cv.fit_transform(movies['tags']).toarray()
```

```python
vector.shape
```

```python
# now we are ploting the graph and and calculating the cosine value 
from sklearn.metrics.pairwise import cosine_similarity 
similarity=cosine_similarity(vector)
print(similarity)
print(movies[movies['title']=='Avatar'])
```

```python
# now i have to define the function and kextract the top 5 similar movies
similarity[0]

def recommand(movie):
    index=movies[movies['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommand('Avatar')
movies

```
